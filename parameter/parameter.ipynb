```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Erste Probemessung
Aloe_Data = "moist_measurement_1.csv"

t_Data = []
moist_Data = []

with open(Aloe_Data, 'r') as csvfile:
    # csv objekt erstellen
    csvreader = csv.reader(csvfile)
    
     # extracting field names through first row
    fields = next(csvreader)
 
 # Daten Zeile f√ºr Zeile einlesen
    for row in csvreader:
        t_Data.append(float(row[0])) 
        moist_Data.append(float(row[1]))
        
        
plt.plot(t_Data, moist_Data)
plt.show()
```
